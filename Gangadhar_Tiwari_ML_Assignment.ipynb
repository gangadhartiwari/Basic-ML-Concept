```python
# Q.14 Convert NaN value using SimpleImputer class  if required? 

sm = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
dataset[['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home']] = sm.fit_transform(dataset[['Teacher_Quality', 
                                                                                                           'Parental_Education_Level', 
                                                                                                           'Distance_from_Home']])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
# Q.13 Convert categorical data using proper encoder type?  

# Define column types
ordinal_cols = [
    "Parental_Involvement", "Access_to_Resources", "Motivation_Level",
    "Family_Income", "Teacher_Quality", "Parental_Education_Level", "Distance_from_Home"]

ordinal_categories = [
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["High School", "College", "Postgraduate"],
    ["Near", "Moderate", "Far"]]

nominal_cols = [
    "Extracurricular_Activities", "Internet_Access", "School_Type",
    "Peer_Influence", "Learning_Disabilities", "Gender"]

column_transformer = ColumnTransformer(transformers=[
    ('ord', OrdinalEncoder(categories=ordinal_categories), ordinal_cols),
    ('nom', OneHotEncoder(drop='first', sparse_output=False), nominal_cols)], remainder="passthrough")
```

```python
# Q.15 Apply Linear Regression Algorithm  to predict Exam_Score? 

x = dataset.iloc[:, :-1]
y = dataset['Exam_Score']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Transform the data
x_train_transformed = column_transformer.fit_transform(x_train)
x_test_transformed = column_transformer.transform(x_test)
```

```python
le = LinearRegression()
le.fit(x_train_transformed, y_train)
```

```python
le.predict(x_test_transformed)
```

```python
le.score(x_train_transformed, y_train)*100
```
