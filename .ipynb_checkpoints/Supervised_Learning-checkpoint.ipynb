<H2 style="color:yellow">LINEAR REGRESSION</H2>

```python
# learning rate (alpha value)
# cost or lost value (function) 
```

```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
```

```python
data = {
    "Dosa(d)": [8, 10, 12, 14, 16], 
    "prize": [10, 13, 16, 19, 22]
}

df = pd.DataFrame(data)
df
```

```python
x = df[["Dosa(d)"]]
```

```python
y = df['prize']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.3, random_state=2)
```

```python
xtrain
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
lr.predict([[18]])
```

<h3>Missiing values</h3>

```python
import numpy as np 
import pandas as pd 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv("DATA1.csv")
df
```

```python
df.columns
```

```python
sm = SimpleImputer(missing_values=np.nan, strategy='mean') 
# it is a special type of function called constructor (it will  get memory allocation for object ). 
# in strategy for mode use can just write "mide_frequent"
```

```python
# if you want to replace nan with an particular value then use fill_value instead of strategy

df[['Age', 'Salary(K)']] = sm.fit_transform(df[['Age', 'Salary(K)']])
```

```python
df
```
